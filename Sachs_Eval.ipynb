```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
seeds = list(range(100,110))
models = ['tabsyn','stasy','tabddpm','codi','great','ctgan','tvae'] 
np.random.seed(0)
```

```python
df_real = pd.read_csv("sachs_train.csv")
# df_real = df_real[df_real["PKC"]>23.5] 
df_test = pd.read_csv("sachs_test.csv")
X_real, y_real = df_real.drop(["P38"],axis=1).values, df_real["P38"].values
X_test, y_test = df_test.drop(["P38"],axis=1).values, df_test["P38"].values
```

```python
res_dict = {"seed": [], "model": [], "r2": [], "mae": [], "mse": []}

for m in models:
    for s in seeds:   
        print(m)
        print(s)
        if m == "real":
            df = df_real
            X, y = df.drop(["P38"],axis=1).values, df["P38"].values
            X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.1, random_state=0)
            clf = XGBRegressor()
        else:    
            df = pd.read_csv(f"synthetic/sachs/{s}/{m}.csv")
            # if len(df[df["PKC"]>23.5]) > 0:
            #     df[df["PKC"]>23.5] 
            #     df = pd.concat([df[df["PKC"]>23.5].head(len(df[df["PKC"]>23.5] )),df[df["PKC"]<=23.5].head(10000-len(df[df["PKC"]>23.5]))])
            # else:
            #     df = df.head(10000)
            # df = df[df["PKC"]>23.5]     
            X_train, y_train = df.drop(["P38"],axis=1).values, df["P38"].values
            # X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.8, random_state=0)
            # print(len(X_train))
            # clf = XGBRegressor()
            # if len(X_train) > 0:
            #     X_train = np.concatenate([X_train, X_real])
            #     y_train = np.concatenate([y_train, y_real])
            # else:
            #     X_train = X_real
            #     y_train = y_real
        clf.fit(X_train, y_train, eval_set=[(X_real, y_real)])
        pred = clf.predict(X_test)
        res_dict["seed"].append(s)
        res_dict["model"].append(m)
        res_dict["r2"].append(r2_score(pred, y_test))
        res_dict["mae"].append(mean_squared_error(pred, y_test))
        res_dict["mse"].append(mean_absolute_error(pred, y_test))
        
```

```python
res_df = pd.DataFrame.from_dict(res_dict)
```

```python
sub_res_df = res_df
# sub_res_df = sub_res_df[sub_res_df["seed"]!=109]
# sub_res_df = sub_res_df[sub_res_df["seed"]!=106]
```

```python
stat_dict = {"model": [], "r2_mean": [], "r2_std": [], "mae_mean": [], "mae_std": [], "mse_mean": [],  "mse_std": []}
metrics = ["r2", "mae", "mse"]
for m in models:
    model_dict = {}
    model_df = sub_res_df.loc[sub_res_df["model"]==m]
    stat_dict["model"].append(m)
    for metric in metrics:
        stat_dict[f"{metric}_mean"].append(model_df[metric].mean())
        stat_dict[f"{metric}_std"].append(model_df[metric].std())
```

```python
stat_df = pd.DataFrame.from_dict(stat_dict)
```
